```python
import cv2
import mediapipe as mp

# ğŸš€ Step 1: Initialize MediaPipe Models
mp_hands = mp.solutions.hands
mp_face = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)
face_detection = mp_face.FaceDetection(min_detection_confidence=0.5)

# ğŸš€ Step 2: Start Real-Time Video Capture
cap = cv2.VideoCapture(0)  # 0 for default webcam

if not cap.isOpened():
    print("Error: Cannot access webcam.")
    exit()

# ğŸš€ Step 3: Real-Time Detection Loop
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Couldn't capture frame.")
        break

    # Convert frame to RGB (MediaPipe expects RGB input)
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect Hands
    hand_results = hands.process(image_rgb)

    # Detect Face
    face_results = face_detection.process(image_rgb)

    # Draw hand landmarks
    if hand_results.multi_hand_landmarks:
        for hand_landmarks in hand_results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    # Draw face bounding box
    if face_results.detections:
        for detection in face_results.detections:
            bboxC = detection.location_data.relative_bounding_box
            h, w, _ = frame.shape
            x, y, w_box, h_box = int(bboxC.xmin * w), int(bboxC.ymin * h), int(bboxC.width * w), int(bboxC.height * h)

            # Draw rectangle around face
            cv2.rectangle(frame, (x, y), (x + w_box, y + h_box), (0, 255, 0), 2)
            cv2.putText(frame, "Face", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # ğŸš€ Step 4: Show Output in Real-Time
    cv2.imshow("Real-Time Face & Hand Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ğŸš€ Step 5: Cleanup
cap.release()
```
